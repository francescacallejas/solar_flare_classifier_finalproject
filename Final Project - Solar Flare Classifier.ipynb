```python
import pandas as pd 

def create_data(input_data):
    file = open(input_data,"r") #open file
    
    data = pd.read_table(file, delim_whitespace=True); 
  
    file.close() #close file
    return data

data = create_data("flare.data1.txt")
```

```python
import sklearn.preprocessing
from sklearn_pandas import DataFrameMapper
from sklearn.model_selection import train_test_split

def split_data(data):
     #transform data
    mapper = DataFrameMapper([
    (['z-class'], sklearn.preprocessing.LabelBinarizer()), 
    (['largest-spot'], sklearn.preprocessing.LabelBinarizer()),
    (['spot-distribution'], sklearn.preprocessing.LabelBinarizer()),
    (['activity'], sklearn.preprocessing.StandardScaler()),
    (['evolution'], sklearn.preprocessing.StandardScaler()), 
    (['day-activity'], sklearn.preprocessing.StandardScaler()),
    (['hist-complex'], sklearn.preprocessing.StandardScaler()),
    (['sun-disk'], sklearn.preprocessing.StandardScaler()),
    (['area'], sklearn.preprocessing.StandardScaler()),
    (['area-spot'], sklearn.preprocessing.StandardScaler()),
    (['c'], sklearn.preprocessing.StandardScaler()),
    (['m'], sklearn.preprocessing.StandardScaler()),
    (['x'], sklearn.preprocessing.StandardScaler())])
    
    """classification_map = {
                          '0': 0,
                          '1': 1,
                          '2': 2,
                          '4': 4
                          }"""
    
    #df = mapper.fit_transform(data)
    
    #split data into training and testing data
    x = data[data.columns.drop('c').drop('m').drop('x')]
    y = data['c']
   
    x_train, x_test, y_train, y_test = train_test_split(x, y)
    return (x_train, x_test, y_train, y_test, mapper)
```

```python
(x_train, x_test, y_train, y_test, mapper) = split_data(data)
```

```python
from sklearn.pipeline import Pipeline

def fit_data(mlp, mapper, x_train, y_train):
    pipeline = Pipeline([
    ('mapper', mapper),
    ('classifier', mlp)
    ])

    model = pipeline.fit(X=x_train, y=y_train)
    
    return model
```
